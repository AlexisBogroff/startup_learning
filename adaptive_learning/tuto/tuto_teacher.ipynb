## Use cases

### Create an exam with its settings, questions and answers

```python
exam = Exam()
exam.set_properties_from_input_main()

questions_ids = []
new_question = True
new_answer = True

question = Question()

# Create content
while new_question:
    question.create()
    while new_answer:
        question.add_answer()
        new_answer = input("Create new answer? (y/N)")
    question.save()
    questions_ids.append(question.id)
    new_question = input("Create new question? (y/N)")

# Add content to exam
for q_id in questions_ids:
    exam.add_question(question.id)

# Show and save exam
exam.show()
exam.save()
```
